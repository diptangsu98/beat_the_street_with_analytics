## Consider a subset ( 30 rows) of @bar. How do 'bidprice', 'askprice', 'average_price' of any stock ( any symbol) are moving in that subset period? Maybe you can try for ‘2020-08-05’. Graphical and data analysis. Dataset @quote @bar dataset

```python
df1=pd.read_csv('bar.csv')
```

```python
df1.head()
```

```python
df2=pd.read_csv('quote.csv')
```

```python
df2.head()
```

##### Filtration of tables based on time column 2020-08-05

```python
temp_df1=df1[df1['time'].str.lower().str.contains('2020-08-05')]
temp_df2=df2[df2['time'].str.lower().str.contains('2020-08-05')]
```

##### Merging of two dataframes into one based on time

```python
dataframe1=pd.merge(temp_df1, temp_df2, on='time')
```

```python
dataframe1.info()
```

```python
dataframe1=dataframe1.tail(30)
```

```python
dataframe1
```

##### Visualizing bid_price, ask_price, average_price for MRK symbol (stock) based on time

```python
plt.figure()
df=dataframe1[['bid_price','ask_price','average_price']].plot(figsize=(14, 6), marker='o', grid=True, markersize=4)
plt.xlabel('Time')
plt.ylabel('Price')
```

## make a table [symbol,ratingBuy,ratingScaleMark,mean of average_price,consensusStartDate,consensusEndDate] from @rating and @bar dataset

```python
df3=pd.read_csv('rating.csv')
```

```python
df3.head()
```

```python
dataframe2=pd.merge(df3, df1, how='inner', on='symbol', left_on=None, right_on=None, left_index=False, right_index=False, sort=True, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
```

```python
dataframe2[['symbol','ratingBuy','ratingScaleMark','consensusStartDate','consensusEndDate','average_price']]
```

## 'average_price' column @bar dataset. How average price changes over 'consensusStartDate' an 'consensusEndDate' from @rating dataset.Graph

```python
plt.figure(figsize=(40,30))
plt.subplots_adjust(hspace = 0.5)
plt.subplot(2,1,1)
sns.lineplot(data=dataframe2, x="consensusStartDate", y="average_price", marker='o', color='blue',linestyle=':')
plt.xticks(rotation=90)
plt.xlabel('consensusStartDate',fontsize=20)
plt.ylabel('average_price',fontsize=20)
plt.subplot(2,1,2)
sns.lineplot(data=dataframe2, x='consensusEndDate', y='average_price', marker='^', color='green', markersize=8, linestyle='dashed')
plt.xticks(rotation=90)
plt.xlabel('consensusEndDate', fontsize=20)
plt.ylabel('average_price', fontsize=20)
```
